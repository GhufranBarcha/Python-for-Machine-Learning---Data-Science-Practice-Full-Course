```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sample = np.random.normal(loc = 50 ,scale = 13 ,size = (1,100)).flatten()
sample = np.round(sample ,0)
sample
```

```python
sns.histplot(sample ,bins = 20 ,kde = True)
```

```python
sns.boxplot(sample)
```

```python
ser = pd.Series(sample)
ser.describe()
```

```python
IQR = 57 - 42
```

```python
lowerlimit =42 -  IQR*1.5
lowerlimit
```

```python
Upperlimit =57 + IQR*1.5
Upperlimit
```

```python
ser1 = ser[ser > lowerlimit]
ser2 = ser[ser < Upperlimit]
```

```python
sns.boxplot(ser1)
```

```python
sns.boxplot(ser2)
```
