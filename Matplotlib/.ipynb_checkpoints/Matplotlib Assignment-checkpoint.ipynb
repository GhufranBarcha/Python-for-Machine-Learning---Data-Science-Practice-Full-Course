# Q1 . Create a E = mc2 formula by using two list

```python
import numpy as np
import matplotlib.pyplot as plt


m = np.linspace(0,10,11)
m
```

```python
c= 3* 10**8
E = m*(c**2)
E
```

```python
## Plot the E = mc2

fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(m,E ,lw = 3 ,color = "red")
axes.set_xlim(0,10)
axes.set_ylim(0)
axes.set_title("E = mc^2")
axes.set_xlabel("M in mass")
axes.set_ylabel("Energy in joules")
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(m,E ,lw = 3 ,color = "red")
axes.set_xlim(0,10)
axes.set_title("E = mc^2")
axes.set_xlabel("M in mass")
axes.set_ylabel("Energy in joule")
plt.yscale("log")
plt.grid(which='both' ,axis = "y")
```
