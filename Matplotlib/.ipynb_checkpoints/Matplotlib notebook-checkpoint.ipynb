```python
## Importing necessary libraries

import matplotlib.pyplot as plt
import numpy as np
```

# 2. Matplotlib Basics 

```python
x = np.arange(0,10)
x
```

```python
y = 2*x
y
```

```python
## Most basic plot
plt.plot(x,y)
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.title("Title linear")
plt.xlim(0,10) # Will set x scale limit
plt.ylim(0,10)  # Will set y scale limit
plt.show()   # will show plot in .py file not necessary in this files
plt.savefig("imagepath.png")   ## will save the image in the path provided
```

## 4. Matplotlib - Implementing Figure and Axis Part two

```python
a = np.linspace(0,10,11)
a
```

```python
b = a**4
b
```

```python
x= np.arange(0,10)
x
```

```python
y = 2*x
y
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(a,b)
plt.show()
```

```python
## Multiple axes

fig = plt.figure()
# Large plot
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(x,y)
axes1.set_title("X and Y")
axes1.set_xlabel("X Xiso")
axes1.set_ylabel("y Xiso")
axes1.set_xlim(0,10)
axes1.set_ylim(0,20)

## small plot
axes2 = fig.add_axes([0.2,0.7,0.2,0.2])
axes2.plot(a,b)
axes2.set_xlim(0,10)
axes2.set_ylim(0,8000)
axes2.set_title("A & B")
axes2.set_xlabel("A")
axes2.set_ylabel("B")
```

```python
## Zooming a plot sextion in other plot

fig = plt.figure()
# Large plot
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
axes1.set_title("X and Y")
axes1.set_xlabel("X Xiso")
axes1.set_ylabel("y Xiso")


## small plot
axes2 = fig.add_axes([1.2,0.8,0.2,0.2])
axes2.plot(a,b)
axes2.set_xlim(4,6)
axes2.set_ylim(0,2000)
axes2.set_title("A & B")
axes2.set_xlabel("A")
axes2.set_ylabel("B")
```
