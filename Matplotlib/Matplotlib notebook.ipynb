```python
## Multiple axes

fig = plt.figure()
# Large plot
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(x,y)
axes1.set_title("X and Y")
axes1.set_xlabel("X Xiso")
axes1.set_ylabel("y Xiso")
axes1.set_xlim(0,10)
axes1.set_ylim(0,20)

## small plot
axes2 = fig.add_axes([0.2,0.7,0.2,0.2])
axes2.plot(a,b)
axes2.set_xlim(0,10)
axes2.set_ylim(0,8000)
axes2.set_title("A & B")
axes2.set_xlabel("A")
axes2.set_ylabel("B")
```

```python
## Zooming a plot sextion in other plot

fig = plt.figure()
# Large plot
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
axes1.set_title("X and Y")
axes1.set_xlabel("X Xiso")
axes1.set_ylabel("y Xiso")


## small plot
axes2 = fig.add_axes([1.2,0.8,0.2,0.2])
axes2.plot(a,b)
axes2.set_xlim(4,6)
axes2.set_ylim(0,2000)
axes2.set_title("A & B")
axes2.set_xlabel("A")
axes2.set_ylabel("B")
```

```python
fig ,axes = plt.subplots(nrows = 1 ,ncols = 3 ,figsize = (10,4))
fig.suptitle("Main haindlae")
for ax in axes:
    ax.plot(a,b)
    
plt.tight_layout()    
```

```python
fig ,axes = plt.subplots(nrows = 2 ,ncols = 2 ,figsize = (8,8))
fig.suptitle("Main haindlae")

axes[0][0].plot(a,b)
axes[0][0].set_xlabel("E axis")
axes[0][0].set_ylabel("Y axis")
axes[0][0].set_title("Title axis")

axes[0][1].plot(x,y)



axes[1][0].plot(a,b)
axes[1][0].set_xlabel("12 axis")
axes[1][0].set_ylabel("213 axis")
axes[1][0].set_title("hjas Title axis")




axes[1][1].plot(x,y)

fig.tight_layout()
```

```python
## Manually set karna ha layout
fig ,axes = plt.subplots(nrows = 2 ,ncols = 2 ,figsize = (8,8) ,dpi = 200)
fig.suptitle("Main haindlae")

axes[0][0].plot(a,b)
axes[0][0].set_xlabel("E axis")
axes[0][0].set_ylabel("Y axis")
axes[0][0].set_title("Title axis")

axes[0][1].plot(x,y)



axes[1][0].plot(a,b)
axes[1][0].set_xlabel("12 axis")
axes[1][0].set_ylabel("213 axis")
axes[1][0].set_title("hjas Title axis")




axes[1][1].plot(x,y)

fig.subplots_adjust(wspace = 0.2 ,hspace = 0.3)
fig.savefig("figplot.png",bbox_inches = "tight")
```

## 8. Matplotlib Styling 

```python
x = np.linspace(0,10,11)
x
```

```python
y = 2*x
y
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,x ,color = "red" ,label = "red")
ax.plot(x,y ,color = "#292929" ,label = "yellow")
ax.plot(x,y+1 ,color = "blue" ,label = "hello")
plt.legend()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,x ,color = "red" ,lw = 13)
ax.plot(x,y ,color = "#292929" ,linewidth = 6)
ax.plot(x,y+1 ,color = "blue" ,linestyle = ":")

ax.plot(x+2,y+1 ,color = "blue" ,linestyle = "-.")
```

```python
## Custom linestyle
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
lines = ax.plot(x ,x+1 ,color = "purple" ,linewidth = 5)
lines[0].set_dashes([5,2,2,2])
```

```python
## Markers

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x ,x+1 ,color = "purple" ,linewidth = 5 ,ls= "--" ,marker = "o" ,markersize = 20)
ax.plot(x+5 ,x+1 ,color = "red" ,linewidth = 5 ,ls= "-" ,marker = "*" ,markersize = 20, markerfacecolor = "black" ,markeredgewidth = 8)
```
