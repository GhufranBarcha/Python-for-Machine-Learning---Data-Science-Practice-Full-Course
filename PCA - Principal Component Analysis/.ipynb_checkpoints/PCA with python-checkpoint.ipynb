```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('33912194-cancer-tumor-data-features.csv')
df
```

## Manual Construction of PCA


### Scaling Data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(df)
```

```python
scaled_X
```

```python
# Because we scaled the data, this won't produce any change.
# We've left if here because you would need to do this for unscaled data
scaled_X - scaled_X.mean(axis=0)
```

```python
# Grab Covariance Matrix
covariance_matrix = np.cov(scaled_X, rowvar=False)
```

```python
# Get Eigen Vectors and Eigen Values
eigen_values, eigen_vectors = np.linalg.eig(covariance_matrix)
```

```python
# Choose som number of components
num_components=2
```

```python
# Get index sorting key based on Eigen Values
sorted_key = np.argsort(eigen_values)[::-1][:num_components]
```

```python
# Get num_components of Eigen Values and Eigen Vectors
eigen_values, eigen_vectors = eigen_values[sorted_key], eigen_vectors[:, sorted_key]
```

```python
# Dot product of original data and eigen_vectors are the principal component values
# This is the "projection" step of the original points on to the Principal Component
principal_components=np.dot(scaled_X,eigen_vectors)
```

```python
principal_components
```

```python
plt.figure(figsize=(8,6))
plt.scatter(principal_components[:,0],principal_components[:,1])
plt.xlabel('First principal component')
plt.ylabel('Second Principal Component')
```

```python
from sklearn.datasets import load_breast_cancer
```

## Implementation in Scikit learn

```python
from sklearn.decomposition import PCA

pca_model = PCA(n_components = 2)
```

```python
# help(PCA)
```

```python
pca_model.fit(scaled_X)
```

```python
pca_model.transform(scaled_X)

```

```python
from sklearn.datasets import load_breast_cancer
datadict = load_breast_cancer()
```

```python
sns.scatterplot(x =pca_model.transform(scaled_X)[:,0],y = pca_model.transform(scaled_X)[:,1] ,hue = datadict["target"])
```

## Fitted Model Attributes

```python
pca_model.components_
```

```python
pca_model.explained_variance_
```

```python
pca_model.explained_variance_ratio_
```

```python
df_comp = pd.DataFrame(pca_model.components_,index=['PC1','PC2'],columns=df.columns)
```

```python
df_comp
```

```python
plt.figure(figsize=(20,3),dpi=150)
sns.heatmap(df_comp,annot=True)
```

```python
np.sum(pca_model.explained_variance_ratio_)
```

```python
pca_30 = PCA(n_components=30)
pca_30.fit(scaled_X)
```

```python
np.sum(pca_30.explained_variance_ratio_)
```

```python
explained_variance = []

for n in range(1,30):
    pca = PCA(n_components=n)
    pca.fit(scaled_X)
    
    explained_variance.append(np.sum(pca.explained_variance_ratio_))
```

```python
plt.plot(range(1,30),explained_variance)
plt.xlabel("Number of Components")
plt.ylabel("Variance Explained");
```
