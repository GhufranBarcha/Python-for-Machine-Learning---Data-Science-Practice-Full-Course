```python
import random
import string

def generate_unique_ids(count=244):
  """Generates an array of unique 5-digit IDs that contain letters.

  Args:
    count: The number of IDs to generate.

  Returns:
    An array of unique 5-digit IDs that contain letters.
  """

  ids = []
  while len(ids) < count:
    new_id = ''.join(random.choices(string.ascii_uppercase + string.digits, k=5))
    if new_id not in ids:
      ids.append(new_id)
  return ids

# Generate an array of 244 unique 5-digit IDs that contain letters.
ids = generate_unique_ids()

# Print the array of IDs.
print(ids)
```

### Left & Right Merge - 18

```python
df1 = pd.DataFrame({'lkey': ['foo', 'bar', 'baa', 'foaa'],
                    'value': [1, 2, 3, 5]})
df2 = pd.DataFrame({'rkey': ['foo', 'bar', 'baz', 'foo'],
                    'value': [5, 3, 7, 8]})
```

```python
## This is left merge
pd.merge(df1 ,df2 ,how = "left" ,on="value")
```

```python
## This is Right merge
pd.merge(df1 ,df2 ,how = "right" ,on="value")
```

### Outer join

```python
## This is take all the values
pd.merge(df1 ,df2 ,how = "outer" ,on="value")
```

```python
import pandas as pd
import sqlite3
# Create a sample DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35]}

df = pd.DataFrame(data)

# Create an in-memory SQLite database and write the DataFrame to it
connection = sqlite3.connect(':memory:')
df.to_sql('my_table', connection, index=False)

# Query the database
query = "SELECT * FROM my_table"
result_df = pd.read_sql(query, connection)

# Print the result
print(result_df)
```
